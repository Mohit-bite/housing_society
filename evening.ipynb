```python
housing=pd.read_csv('housing_data.csv')
```

```python
housing.info()
```

```python
housing.head()
```

```python
housing['period_begin']=pd.to_datetime(housing['period_begin'])
housing['period_end']=pd.to_datetime(housing['period_end'])
housing['region_name']=housing['region_name'].astype('category')
housing['inventory']=housing['inventory'].fillna(0).astype("int32")
```

```python
housing.head()
```

```python
ca_housing=housing[housing['region_name'].str.contains('CA')]
ca_housing['year']=ca_housing['period_begin'].dt.year
ca_housing
```

```python
ca_group=ca_housing.groupby('year')[['median_active_list_price']].mean()
ca_group.sort_values(by='median_active_list_price')
```

```python
fig,ax=plt.subplots()
ax.plot(ca_group.index,ca_group.values,linestyle="--",color='red')
fig.suptitle(" Ca Price Graph")
plt.show()
```
